```python
save_counter = 0
crop_counter = 0
folder_path = 'DATASET_1\\ImposterRaw\\'
saved_folder_path = 'DATASET_1\\ImposterRawSaved\\'
cropped_folder_path = 'ImposterRawTestFinal\\'
```

```python
if not os.path.exists(saved_folder_path):
    os.mkdir(saved_folder_path)
```

```python
if not os.path.exists(cropped_folder_path):
    os.mkdir(cropped_folder_path)
```

```python
for line in lines:
    raw_image_path = folder_path+line.strip()
    raw_image = cv2.imread(raw_image_path)
    cv2.imwrite(os.path.join(saved_folder_path, 'face'+str(save_counter)+'.jpg'), raw_image)
    save_counter+=1
```

# Saved images in a folder and iterating through it

```python
for images in os.listdir(saved_folder_path):
    absolute_path = saved_folder_path + images
    image_color = cv2.cvtColor(cv2.imread(absolute_path), cv2.COLOR_RGB2BGR)
    detections = detector.detect_faces(image_color)
    image_copy = image_color.copy()
    confidence = 0
    for det in detections:
        if(det['confidence'] > confidence):
            confidence = det['confidence']
            x, y, width, height = det['box']
            cropped_image = image_copy[y:y+height, x:x+width]       
    cv2.imwrite(os.path.join(cropped_folder_path , 'face'+str(crop_counter)+'.jpg'),cropped_image)
    crop_counter+=1
```
