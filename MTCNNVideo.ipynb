```python
import os
```

```python
import cv2
```

```python
from mtcnn.mtcnn import MTCNN
```

```python
import matplotlib.pyplot as plt
```

```python
detector = MTCNN()
```

```python
test_path = 'C://Users//PriyankaPrabhu//FaceDetection//DATASET3//test//attack//hand//'
output_path = 'DATASET3//output//'
counter = 0
```

```python
for videos in os.listdir(test_path):
    absolute_path = test_path + videos
    print(absolute_path)
    video = cv2.VideoCapture(absolute_path)
    frame_width = int(video.get(3))
    frame_height = int(video.get(4))
    size = (frame_width, frame_height)
    result = cv2.VideoWriter(output_path+'filename.avi',cv2.VideoWriter_fourcc(*'MJPG'),29, size)
    frame_num = 0
    confidence = 0
    while(True):
        ret, frame = video.read()
        frame_num += 1
        print(ret)
        if ret == True:
            detections = detector.detect_faces(frame)
            if len(detections) > 0:
                for det in detections:
                    if(det['confidence'] > confidence):
                        x, y, width, height = det['box']
                        x2, y2 = x + width, y + height
                        image_copy = cv2.rectangle(frame, (x, y), (x2, y2), (0, 0, 255), 4)
                plt.figure(figsize = (10,10))
                cv2.imwrite(os.path.join('SampleVideoOutput\\' , 'face'+str(counter)+'.jpg'),image_copy)
                counter+=1
        else:
            break
```
