```python
from mtcnn import MTCNN
```

```python
import cv2
```

```python
import os
```

```python
import matplotlib.pyplot as plt
```

```python
detector = MTCNN()
```

```python
f = open('DATASET_1/client_test_raw.txt', 'r')
```

```python
f
```

```python
lines = f.readlines()
```

```python
f.close()
```

```python
lines
```

```python
counter = 0
folder_path = 'DATASET_1\\ClientRaw\\'
```

```python
for line in lines:
    Image_path = folder_path+line.strip()
    image_color = cv2.cvtColor(cv2.imread(Image_path), cv2.COLOR_RGB2BGR)
    detections = detector.detect_faces(image_color)
    image_copy = image_color.copy()
    confidence = 0
    for det in detections:
        if(det['confidence'] > confidence):
            confidence = det['confidence']
            x, y, width, height = det['box']
            x=210
            keypoints = det['keypoints']
            cv2.rectangle(image_copy, (x,y), (x+width,y+height), (0,155,255), 2)
    plt.figure(figsize = (10,10))
    cv2.imwrite(os.path.join('ClientRawTest\\' , 'face'+str(counter)+'.jpg'),image_copy)
    counter+=1
```
